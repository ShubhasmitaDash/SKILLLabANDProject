```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris

from sklearn.metrics import mean_absolute_error, mean_squared_error

class MultipleLinearRegression:
    def __init__(self):
        self.intercept = None
        self.coef = None

    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1) # add a column of ones for the intercept term
        XT_X = np.dot(X.T, X)
        XT_y = np.dot(X.T, y)
        self.coef = np.linalg.solve(XT_X, XT_y)
        self.intercept = self.coef[0]
        self.coef = self.coef[1:]

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        return np.dot(X, np.concatenate(([self.intercept], self.coef)))

# Load dataset
X = np.array([[1, 2], [2, 4], [3, 5], [4, 6], [5, 8]])
y = np.array([2, 4, 5, 4, 5])

# Train the model
model = MultipleLinearRegression()
model.fit(X, y)

# Print the coefficients
print("Intercept: ", model.intercept)
print("Coefficients: ", model.coef)

# Make predictions
X_new = np.array([[6, 10], [7, 11], [8, 13]])
y_pred = model.predict(X_new)


# Plot the data and plane of best fit
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], y)
xx, yy = np.meshgrid(np.linspace(0, 8, 10), np.linspace(0, 14, 10))
zz = model.intercept + model.coef[0] * xx + model.coef[1] * yy
ax.plot_surface(xx, yy, zz, alpha=0.2)
ax.set_xlabel('Sepal length')
ax.set_ylabel('Sepal width')
ax.set_zlabel('Petal width')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

class MultipleLinearRegression:
    def __init__(self):
        self.intercept = None
        self.coef = None

    def fit(self, X, y):
        X = np.insert(X, 0, 1, axis=1) # add a column of ones for the intercept term
        XT_X = np.dot(X.T, X)
        XT_y = np.dot(X.T, y)
        self.coef = np.linalg.solve(XT_X, XT_y)
        self.intercept = self.coef[0]
        self.coef = self.coef[1:]

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        return np.dot(X, np.concatenate(([self.intercept], self.coef)))

# Load iris dataset
iris = load_iris()
X = iris.data[:, :3] # sepal length, sepal width, petal length, petal width
y = iris.data[:, 3] # petal width

# split the dataset into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 3)

# Train the model
model = MultipleLinearRegression()
model.fit(X_train, y_train)

# Print the coefficients
print("Intercept: ", model.intercept)
print("Coefficients: ", model.coef)

# Make predictions

y_pred = model.predict(X_test)

#Calculate Mean absolute error and Mean squared error for test data
mae = mean_absolute_error(y_pred, y_test)
mse = mean_squared_error(y_pred, y_test)
print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)



```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ------------------------------
# 2. Load the Dataset
# ------------------------------

# Load iris dataset
iris = load_iris()
X = iris.data[:, :3] # sepal length, sepal width, petal length, petal width
y = iris.data[:, 3] # petal width

# split the dataset into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 3)

# ------------------------------
# 2. Create and train model
# ------------------------------
model = LinearRegression()
model.fit(X_train, y_train)

# ------------------------------
# 3. Get learned parameters
# ------------------------------
print("Intercept (b0):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)

# ------------------------------
# 4. Make predictions
# ------------------------------
y_pred = model.predict(X_test)
print("\nPredictions:", y_pred)

# ------------------------------
# 5. Evaluate the model
# ------------------------------
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation:")
print("Mean Squared Error:", mse)
print("R Score:", r2)
```
